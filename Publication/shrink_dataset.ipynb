# Filter Time

```python
year = 2017
years_new =['2017', '2018', '2019', '2020', '2021', '2022']
data = data_raw[data_raw.index.year >= year]
data
```

```python
def filter_data(df, stations):
    filtered_columns = [col for col in df.columns if any(s in col for s in stations)]
    return df[filtered_columns]
```

```python
def save_shrinked_dataset(dataset_old, data_new, stations_new, years_new, filename):
    dataset_new = dataset_old
    dataset_new["data"] = data_new
    dataset_new["stations"]= stations_new
    dataset["years"] = years_new


    filepath = f'../data/datasets/publication/center_42003/{filename}.pickle'
    with open(filepath, 'wb') as f:
        pickle.dump(dataset, f)

    print("File successfully saved:")
    print(filepath)
```

```python
stations_stage_0 = ["42003"]
data_stage_0 = filter_data(data, stations_stage_0)
save_shrinked_dataset(dataset, data_stage_0, stations_stage_0, years_new, "42003_stage_0")
data_stage_0
```

```python
stations_stage_1 = ["42040", "42001", "42003", "41012", "42036", "42039"]
data_stage_1 = filter_data(data, stations_stage_1)
save_shrinked_dataset(dataset, data_stage_1, stations_stage_1, years_new, "42003_stage_1")
data_stage_1
```

```python
stations_stage_2 = ["42040", "42038", "42007", "42001", "42002", "42003", "42012", "42019", "42035", "42036", "42039", "41041", "42055"]
data_stage_2 = filter_data(data, stations_stage_2)
save_shrinked_dataset(dataset, data_stage_2, stations_stage_2, years_new, "42003_stage_2")
data_stage_2
```
