```python
import pickle
```

```python
DATAFILE = "dataset_GOM_1_A_A.pickle"
```

```python
with open(f'../data/datasets/type_A/{DATAFILE}', 'rb') as f:
    # load the object from the file using pickle.load()
    dataset = pickle.load(f)
```

```python
data = dataset["data"]
```

```python
print("Stations: ", dataset["stations"])
print("Years: ", dataset["years"])
print("features: ", dataset["features"])
```

```python
data
```

```python
def filter_data(df, stations):
    filtered_columns = [col for col in df.columns if any(s in col for s in stations)]
    return df[filtered_columns]
```

```python
def save_shrinked_dataset(dataset_old, data_new, filename):
    dataset_new = dataset_old
    dataset_new["data"] = data_new

    filepath = f'../data/datasets/publication/{filename}.pickle'
    with open(filepath, 'wb') as f:
        pickle.dump(dataset, f)

    print("File successfully saved:")
    print(filepath)
```

```python
stations_stage_0 = ["42040"]
data_stage_0 = filter_data(data, stations_stage_0)
save_shrinked_dataset(dataset, data_stage_0, "GOM_stage_0")
data_stage_0
```
